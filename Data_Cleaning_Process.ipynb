# **Cleaning Data: Remove image files having size == 0 Bytes**

```python
# import os
import os 
```

```python
# Initial count of files in each folder 
total = 0
counts = {}
for root, dirs, files in os.walk("/content/data"):
    if not dirs:
        print(f'{root}: {len(files)}')
        counts[root] = len(files)
        total += len(files)
```

```python
# Loop through nested folders 
rootdir = '/content/data'

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        img_path = os.path.join(subdir, file)
        # Check of file size is equal to 0 and if so, remove file
        if os.path.getsize(img_path) == 0:  
          os.remove(img_path)
```

```python
# Check if there are files having 0 byte size after removing 
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        img_path = os.path.join(subdir, file)
        if os.path.getsize(img_path) == 0:
          print(img_path)
```

```python
# Get the count of exact files from each folder after cleaning 
total = 0
counts = {}
for root, dirs, files in os.walk("/content/data"):
    if not dirs:
        print(f'{root}: {len(files)}')
        counts[root] = len(files)
        total += len(files)
```
